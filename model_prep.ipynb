```python
#  Search for Python Repositories with >100 Stars
 # Define your search query for Python repositories with more than 100 stars
query = "language:python stars:>100 size:<500"  # Only repos with size < 500 KB
repos = g.search_repositories(query=query)
 # Print basic information about the first 5 repositories
for repo in repos[:5]:  # Only displaying the top 5 results
     print(f"Repository Name: {repo.full_name}")
     print(f"Stars: {repo.stargazers_count}")
     print(f"Description: {repo.description}\n")
```

```python
import time

code_samples = []
max_repos = 3  # Limit to 3 repositories
max_files = 10  # Limit to 10 .py files
max_depth = 1   # Limit folder depth

def fetch_contents(repo, path="", depth=0):
    if depth > max_depth or len(code_samples) >= max_files:
        return []
    
    try:
        contents = repo.get_contents(path)
        result = []
        
        for file_content in contents:
            if file_content.type == "dir":
                result.extend(fetch_contents(repo, file_content.path, depth + 1))
            elif file_content.name.endswith(".py"):
                result.append({
                    "repo": repo.full_name,
                    "file_name": file_content.path,
                    "code": file_content.decoded_content.decode("utf-8", errors="ignore")
                })
                if len(result) >= max_files:
                    break
        return result
    
    except Exception as e:
        print(f"Error fetching files from {repo.full_name}: {e}")
        return []

# Limit repository search
for repo in repos[:max_repos]:
    if len(code_samples) >= max_files:
        break
    print(f"Fetching contents of repository: {repo.full_name}")
    code_samples.extend(fetch_contents(repo))
    time.sleep(1)  # Delay to avoid rate limiting

print(f"Collected {len(code_samples)} Python files.")
```

```python
# Output the results
if code_samples:
    for sample in code_samples:
        print(f"Repository: {sample['repo']}")
        print(f"File Name: {sample['file_name']}")
        print(f"Code Snippet:\n{sample['code'][:100]}...")  # Print the first 100 characters of the code
        print("-" * 50)
else:
    print("No Python files found in the scraped repositories.")
```

```python
cleaned_samples = []
for sample in code_samples:
     lines = sample["code"].strip().split("\n")
     if len(lines) > 10 and len(lines) <= 1000 and "__init__" not in sample["file_name"]:
         cleaned_samples.append(sample)
```

```python
import pandas as pd
 # Convert the cleaned data to a DataFrame
df = pd.DataFrame(cleaned_samples)
 # Save the code snippets to a CSV file
df.to_csv("github_code_snippets.csv", index=False)
 # Verify saved data
print(df.head())
```

```python
from codecarbon import track_emissions
import pandas as pd

df = pd.read_csv("github_code_snippets.csv")

# Measure energy consumption
# Use @track_emissions to track energy.
@track_emissions()
def run_code(code):
    try:
        exec(code) # Run the code using exec() , which executes Python code.
    except Exception as e:
        print(f"Error running code: {e}")
# Apply energy measurement to each code snippet
df["energy_usage"] = df["code"].apply(lambda x: run_code(x))
df.to_csv("code_energy_dataset.csv", index=False)
```

```python
from codecarbon import EmissionsTracker
import pandas as pd
import time

# Load the code snippets
df = pd.read_csv("github_code_snippets.csv")

# Function to measure energy consumption
def run_code_with_energy_tracking(code):
    tracker = EmissionsTracker()
    tracker.start()
    try:
        exec(code)  # Runs the code snippet
    except Exception as e:
        print(f"Error running code: {e}")
    emissions = tracker.stop()
    return emissions

# Measure energy consumption for each snippet
df["energy_usage"] = df["code"].apply(lambda x: run_code_with_energy_tracking(x))

# Save updated data
df.to_csv("code_energy_dataset.csv", index=False)
```

### Overview
This document explains how we measure the energy consumption of Python code snippets collected from GitHub. The goal is to identify inefficient code that consumes a lot of energy.

### Steps Completed

2. **Energy Measurement Setup**:

   - The `EmissionsTracker` is used to track emissions while the code executes.

3. **Executing Code with Tracking**:
   - The `run_code_with_energy_tracking` function:
     - Starts energy tracking using `EmissionsTracker`.
     - Executes the code snippet safely using `exec()`.
     - Handles any errors to prevent failures.
     - Stops tracking and records the energy usage.

4. **Storing Results**:
   - The measured energy consumption is stored in a new column, `energy_usage`.
   - The updated dataset is saved as `code_energy_dataset.csv` for further analysis.
